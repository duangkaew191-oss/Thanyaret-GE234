## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile


```python

import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/64ee3517-4e16-418f-81da-722c123ddacd/resource/ed6ff6ec-fbe6-43a5-99f1-e057c2796c33/download/th_province_boundary.shp")

# แสดงข้อมูล 5 แถวแรก
print(gdf.head())
```


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python

# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร)
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
print(gdf.nlargest(5, "area_sqkm")[["PROV_NAME", "area_sqkm"]])
```


## 🔹 ตัวอย่างที่ 5: ใช้ GeoPandas ทำ Spatial Join


```python

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/ed8e818c-e1e3-4746-a7cd-6e437c078b9d/resource/9db5a579-2673-4e83-bf4f-bc7a83316c35/download/th_district_boundary.shp")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```

แบบฝึกหัด 1: ใช้ NumPy คำนวณค่า Mean, Max, Min ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

```python
import numpy as np


nir_ex1 = np.array([[0.6, 0.7, 0.5],
                    [0.8, 0.6, 0.4],
                    [0.5, 0.8, 0.3]])

red_ex1 = np.array([[0.2, 0.3, 0.1],
                    [0.4, 0.2, 0.1],
                    [0.1, 0.2, 0.05]])

# คำนวณค่า NDVI
denominator = nir_ex1 + red_ex1
denominator[denominator == 0] = 1e-9 # ป้องกันการหารด้วยศูนย์
ndvi_ex1 = (nir_ex1 - red_ex1) / denominator

print("ค่า NDVI สำหรับแบบฝึกหัด 1:")
print(ndvi_ex1)


print(f"\nค่าเฉลี่ย NDVI: {np.mean(ndvi_ex1):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi_ex1):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi_ex1):.2f}")
```

### **แบบฝึกหัด 2**: ใช้ GeoPandas โหลด Shapefile ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

```python
import geopandas as gpd
import os
import zipfile
import requests

url = "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
local_zip_path = "/tmp/tha_admbnda_adm1_rtsd_20190221.zip"
extracted_dir = "/tmp/tha_admbnda_adm1_rtsd_20190221"

os.makedirs(extracted_dir, exist_ok=True)

print(f"กำลังดาวน์โหลดไฟล์ Shapefile จาก: {url}")

try:
    response = requests.get(url, stream=True)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    with open(local_zip_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"ดาวน์โหลดไฟล์สำเร็จ: {local_zip_path}")
except requests.exceptions.RequestException as e:
    print(f"Error: ไม่สามารถดาวน์โหลดไฟล์จาก {url} ได้: {e}")
    raise


if os.path.exists(local_zip_path) and os.path.getsize(local_zip_path) > 0:
    try:
        with zipfile.ZipFile(local_zip_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"แตกไฟล์ไปที่: {extracted_dir}")

        shp_file_path = None
        for root, dirs, files in os.walk(extracted_dir):
            for file in files:
                if file.endswith(".shp"):
                    shp_file_path = os.path.join(root, file)
                    break
            if shp_file_path:
                break

        if shp_file_path:
            gdf_ex2 = gpd.read_file(shp_file_path)

            print("\nข้อมูล 5 แถวแรกของ Shapefile จังหวัดที่โหลดมา:")
            display(gdf_ex2.head())

            print(f"\nCRS ของข้อมูลจังหวัด: {gdf_ex2.crs}")

            # คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางกิโลเมตร)
            if gdf_ex2.crs.is_geographic: # ตรวจสอบว่าเป็น Geographic CRS (ละติจูด/ลองจิจูด)
                print("\nCRS เป็น Geographic CRS กำลังแปลงเป็น Projected CRS เพื่อคำนวณพื้นที่ที่แม่นยำ...")
                if not gdf_ex2.geometry.is_empty.all():
                    # WGS 84 / UTM zone 47N (EPSG:32647) is suitable for a large part of Thailand
                    gdf_ex2_proj = gdf_ex2.to_crs("EPSG:32647")
                    gdf_ex2["area_sqkm_ex2"] = gdf_ex2_proj.geometry.area / 1e6
                else:
                    print("Warning: Geometry is empty, cannot calculate area.")
                    gdf_ex2["area_sqkm_ex2"] = 0
            else:
                gdf_ex2["area_sqkm_ex2"] = gdf_ex2.geometry.area / 1e6

            print("\nจังหวัด 5 อันดับแรกที่มีพื้นที่มากที่สุด (จากแบบฝึกหัด 2):")
            # Dynamically find a suitable province name column
            province_name_col = None
            for col in ['PROV_NAME', 'ADM1_EN', 'NAME_1']:
                if col in gdf_ex2.columns:
                    province_name_col = col
                    break

            if province_name_col:
                display(gdf_ex2.nlargest(5, "area_sqkm_ex2")[[province_name_col, "area_sqkm_ex2"]])
            else:
                display(gdf_ex2.nlargest(5, "area_sqkm_ex2")[["area_sqkm_ex2"]])
        else:
            print(f"Error: ไม่พบไฟล์ .shp ภายใน {extracted_dir}")

    except zipfile.BadZipFile:
        print(f"Error: ไฟล์ที่ดาวน์โหลดมาไม่ใช่ไฟล์ Zip ที่ถูกต้อง: {local_zip_path}")
    except Exception as e:
        print(f"เกิดข้อผิดพลาดระหว่างแตกไฟล์หรืออ่าน Shapefile: {e}")
else:
    print(f"Error: ไม่สามารถดาวน์โหลดไฟล์จาก {url} ได้สำเร็จ หรือไฟล์ที่ดาวน์โหลดมาว่างเปล่า")
```

แบบฝึกหัด 3: ใช้ GeoPandas ทำ Spatial Join ระหว่างข้อมูลจังหวัดและอำเภอ

```python
import geopandas as gpd
from IPython.display import display

gdf = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/6f054351-bf2c-422e-8deb-0a511d63a315/resource/84b5e1ef-31f7-4760-bc3a-ac58269acc87/download/diaphanhuyen.zip")
print("ข้อมูลดิบ (gdf.head()):")
display(gdf.head())

gdf_districts = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/6f054351-bf2c-422e-8deb-0a511d63a315/resource/84b5e1ef-31f7-4760-bc3a-ac58269acc87/download/diaphanhuyen.zip")

gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

print("ข้อมูลที่ผ่าน Spatial Join (gdf_joined ทั้งหมด):")
display(gdf_joined)
```

แบบฝึกหัด 4: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

```python
import numpy as np
import geopandas as gpd
from IPython.display import display

# ตรวจสอบว่า gdf_ex2 มีอยู่จริงหรือไม่ (ควรจะถูกสร้างในเซลล์ e6ab145c)
# ถ้าไม่มี ให้รันเซลล์ e6ab145c ก่อน
if 'gdf_ex2' not in locals():
    print("Warning: gdf_ex2 is not found. Please run Exercise 2 cell (e6ab145c) first.")
    # As a fallback for demonstration, load a simple version if gdf_ex2 is truly missing
    # In a real scenario, the user would need to run the previous cell.
    try:
        # Attempt to load the same file used in Ex2 if gdf_ex2 is not in locals
        extracted_dir = "/tmp/tha_admbnda_adm1_rtsd_20190221"
        shp_file_path = None
        for root, dirs, files in os.walk(extracted_dir):
            for file in files:
                if file.endswith(".shp"):
                    shp_file_path = os.path.join(root, file)
                    break
            if shp_file_path:
                break
        if shp_file_path:
            gdf_ex2 = gpd.read_file(shp_file_path)
            # Simulate area calculation for consistency if reloaded
            if gdf_ex2.crs.is_geographic:
                gdf_ex2_proj = gdf_ex2.to_crs("EPSG:32647")
                gdf_ex2["area_sqkm_ex2"] = gdf_ex2_proj.geometry.area / 1e6
            else:
                gdf_ex2["area_sqkm_ex2"] = gdf_ex2.geometry.area / 1e6
        else:
            raise FileNotFoundError(f"Cannot find .shp file in {extracted_dir}")
    except Exception as e:
        print(f"Error reloading gdf_ex2 for fallback: {e}")
        # Create a dummy DataFrame if all else fails to allow code execution
        data = {'PROV_NAME': ['Province A', 'Province B', 'Province C'],
                'geometry': [None, None, None],
                'area_sqkm_ex2': [1000, 2000, 1500]}
        gdf_ex2 = gpd.GeoDataFrame(data)


# --- Simulate NDVI data for each province using NumPy ---
# สร้างอาร์เรย์ของค่า NDVI สุ่มหนึ่งค่าสำหรับแต่ละจังหวัดใน gdf_ex2
# ค่า NDVI มักจะอยู่ระหว่าง -1 ถึง 1. เราจะจำลองค่าที่สมจริง (เช่น 0.1 ถึง 0.9 สำหรับพื้นที่พืชพรรณ)
np.random.seed(42) # เพื่อให้ผลลัพธ์สามารถทำซ้ำได้
simulated_ndvi_values = np.random.uniform(0.1, 0.9, size=len(gdf_ex2))

# เพิ่มค่า NDVI จำลองเป็นคอลัมน์ใหม่ใน DataFrame gdf_ex2
gdf_ex2['avg_ndvi'] = simulated_ndvi_values

print("ข้อมูลจังหวัดพร้อมค่า NDVI เฉลี่ยจำลอง (5 แถวแรก):")
display(gdf_ex2.head())

# --- ค้นหาจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุดโดยใช้ GeoPandas/Pandas ---
# ใช้ชื่อคอลัมน์จังหวัดที่ระบุจากแบบฝึกหัดที่ 2, หรือใช้ค่าเริ่มต้นเป็น 'ADM1_EN'
province_name_col_ex4 = None
for col_name in ['PROV_NAME', 'ADM1_EN', 'NAME_1']:
    if col_name in gdf_ex2.columns:
        province_name_col_ex4 = col_name
        break

if province_name_col_ex4:
    province_with_highest_ndvi = gdf_ex2.loc[gdf_ex2['avg_ndvi'].idxmax()]

    print(f"\nจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุดคือ:")
    print(f"ชื่อจังหวัด: {province_with_highest_ndvi[province_name_col_ex4]}")
    print(f"ค่า NDVI เฉลี่ย: {province_with_highest_ndvi['avg_ndvi']:.2f}")

    # แสดงแถวข้อมูลทั้งหมดเพื่อบริบท
    display(province_with_highest_ndvi[[province_name_col_ex4, 'avg_ndvi', 'area_sqkm_ex2']].to_frame().T)
else:
    print("ไม่พบคอลัมน์ชื่อจังหวัดที่เหมาะสมใน DataFrame.")

```
